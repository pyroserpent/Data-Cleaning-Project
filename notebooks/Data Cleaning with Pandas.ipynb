```python
import pandas as pd
import re
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_excel(r"C:\Users\ziaul\OneDrive\Documents\pandas\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
df['Last_Name'] = df['Last_Name'].str.replace(r'[^A-Za-z]', '', regex=True)
df
```

```python
def clean_phone_number(phone):
    if pd.isna(phone) or str(phone).strip().lower() in ["n/a", "nan", "none", ""]:
        return None  # ✅ Handle NaN, "N/a", "None", empty strings
    phone_str = str(phone).strip()  # ✅ Convert to string and remove spaces
    digits = re.sub(r'\D', '', phone_str)  # ✅ Remove all non-numeric characters

    # ✅ Handle country codes (leading "1" for US numbers)
    if len(digits) == 11 and digits.startswith('1'):
        digits = digits[1:]  # Remove leading "1"
    # ✅ Ensure we only keep numbers with exactly 10 digits
    if len(digits) != 10:
        return None  
    # ✅ Format correctly as (XXX) XXX XXXX
    return f"({digits[:3]}) {digits[3:6]} {digits[6:]}"
# Apply function to DataFrame
df['Formatted_Phone'] = df['Phone_Number'].apply(clean_phone_number)

from IPython.display import display

# Display cleaned data
display(df)
```

```python
df['Phone_Number'] = df['Formatted_Phone']

# ✅ Remove the now redundant 'Formatted_Phone' column
df.drop(columns=['Formatted_Phone'], inplace=True)

# ✅ Display cleaned data in Jupyter Lab
display(df)
```
